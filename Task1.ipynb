#Task 1
Create a bar chart or histogram to visualize the distribution of a categorical or continous variable, such as the distribution of ages or genders in a population

```python
#Import libraries
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
import seaborn.objects as so
```

```python
#Read csv that contains female population in every country
fp=pd.read_csv('FemalePopulatiospyr.csv')
f_ppy = fp[['Country Name','2012','2022']]
```

```python
#Reorder values of population in a descending order
fp_top10_2012 = f_ppy.sort_values(by='2012', ascending=False).head(10)
fp_top10_2022 = f_ppy.sort_values(by='2022', ascending=False).head(10)
```

```python
#Set the type pf style of the graph
sns.set(style="ticks")
```

```python
#Plotting the data 
plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=fp_top10_2022, color='blue', label='2022 Population')
sns.barplot(x='2012', y='Country Name', data=fp_top10_2012, color='orange', label='2012 Population')
plt.xlabel('Population')
plt.ylabel('Countries')
plt.title('Top 10 Countries with Most Population of Females (2012 vs 2022)')
plt.legend()
plt.show()
```

```python
#Read csv that contains male population in every country
mp=pd.read_csv('malespopulation.csv')
m_ppy = mp[['Country Name','2012','2022']]
```

```python
#Reorder values of population in a descending order
mp_top10_2012 = m_ppy.sort_values(by='2012', ascending=False).head(9)
mp_top10_2022 = m_ppy.sort_values(by='2022', ascending=False).head(9)
```

```python
#Set the type pf style of the graph
sns.set(style="ticks")
```

```python
#Plotting the data 
plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', data=mp_top10_2022, color='blue', label='2022 Population')
sns.barplot(x='2012', y='Country Name', data=mp_top10_2012, color='orange', label='2012 Population')
plt.xlabel('Population')
plt.ylabel('Countries')
plt.title('Top 10 Countries with Most Population of Males (2012 vs 2022)')
plt.legend()
plt.show()
```

```python
#Read csv that contains male and female population 
dff_mf=pd.read_csv('fema_male.csv')
f_m = dff_mf[['Country Name','2012','2022']]
```

```python
#Set the type pf style of the graph
sns.set(style='ticks')
```

```python
#Plotting the data 
plt.figure(figsize=(10, 6))
sns.barplot(x='2022', y='Country Name', hue='2022',data=f_m, palette='muted',legend=False)
sns.barplot(x='2012', y='Country Name', hue='2012',data=f_m, palette='pastel',legend=False)
plt.xlabel('Population in Mexico')
plt.ylabel('Gender')
plt.title('Population of Male and Female in Mexico (2012 vs 2022)')
plt.show()
```

Conclusions:
- India is the country that had significant growth in men and women from 2012 to 2022
- Japan is the country that did not have significant growth in both men and women from 2012 to 2022.
- The proportion in which the population in Mexico grew, both of women and men, is very similar, although there are more women in the country
